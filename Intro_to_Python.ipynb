# 3.Variables 

Let's load weather data recorded from Chicago O'Hare in 2015. 

````python
numpy.loadtxt('OHARE_TEMP_2015.csv',delimiter=',')
````

`numpy.loadtxt('OHARE_TEMP_2015.csv',delimiter=',')` let's breakdown this command

We called the numpy loadtxt method from numpy by using the syntax *numpy.somecommand*. This is a method from the numpy library. You can think of methods as verbs that do things in your program. In this case this loads data from a CSV file. The method takes arguments or parameters and returns an output. In this case we passed two arguments, `OHARE_TEMP_2015.csv` and `delimter=','` in two different ways, as *positional* and *keyword* arguments, respectively.  

The filename `OHARE_TEMP_2015.csv` is the positional argument that has to be passed to the method first. The keyword argument is the `delimiter` keyword. Our file has fields separated by commas so to pass this information into the function we pass it using the keyword `delimiter` by setting `delimiter` equal to the string `,`, `delimiter=','`'.

When we call the function loadtxt it outputs an array of data. We can capture the output by storing it in a variable that we can reference later. You can think of the variable as a label on the data. In python variables 
must start with a letter. Also a good variable should be self-documenting on what the variable labels. This means `np_ohare_temperature` is better than, say, `x` because the former communicates the variable is labeling a numpy array that contains temperature data from ohare, x communicates nothing. 

````python
np_ohare_temperature = numpy.loadtxt('OHARE_TEMP_2015.csv',delimiter=',')
````

`0:7` is an instruction to start of the 0th element of the list and go up to but not including the 7th array of the list

````python
np_ohare_temperature[:7]
````

If we omit the right index in the slice it will by default end at the end of the array. 

````python
np_ohare_temperature[360:]
````

# Loops and if statements

Loops are control structures that are allow for the repetition of blocks of code of values or until a certain condition is met. We will go over the `for` loop in python. An `if` statement tests a condition that is either `True` or `False`. If the condition is true the code inside the if statement will be executed. 

Suppose we wanted to find the first time that the temperature is in the single digits in Chicago. How would we do it?

One solution is to go through the array one-by-one until a value is found. 

````python
day = np_ohare_temperature[0,1]
if day < 0:
    print('Found it ', day)
day = np_ohare_temperature[1,1]
if day < 0:
    print('Found it ', day)
day = np_ohare_temperature[2,1]
if day < 0:
    print('Found it ', day)
day = np_ohare_temperature[3,1]
if day < 0:
    print('Found it ', day)
day = np_ohare_temperature[4,1]
if day < 0:
    print('Found it ', day)
day = np_ohare_temperature[5,1]
if day < 0:
    print('Found it ', day)
day = np_ohare_temperature[6,1]
if day < 0:
    print('Found it ', day)
day = np_ohare_temperature[7,1]
if day < 0:
    print('Found it', day)
````

As you can see, there is a bit cumbersome and not scalable as you would have to do this manually. One things we can see here is that there is a large amount of repetition until we hurt a certain condition. This is where loops come in handy!

````python
for day in np_ohare_temperature[:,1]:
    if day < 0:
        print('Found it', day)
        break
````

Break down of the loop: 
```
for day in np_ohare_temperature[:,1]:
    if day < 0:
        print('Found it', day)
        break
```

The basic syntax of a for loop the following: 
```
for variable in some_list_or_string:
    #do something with the variable
```
Instead of an array we can also loop over a string:


````python
for i in 'Chicago':
    print(i)
````

In our example we used a slice to get the temperatures, `np_ohare_temperatures[:,1]`. We know this command will return a sub-array of the temperatures. The loop will then loop over this sub-array. In each iteration of the loop the value in the sub-array will be stored in the variable `day`. We then test whether day is less than zero and if the test evalutes to `True` we execute the code in the `for` loop which prints out the value stored in `day`. The `break` command is used to break out of a loop and executed within the if statement. 

## Functions

Another important aspect of the Python language is functions. Functions are important because they encapsulate code into reusable chunks. An example could be converting the temperatures in Farenheight to Celsius. A function is defined as the following:

def some_function(parameter1,parameter2)
    #do something with parameter1 and paratmer2

````python
def farenheit_to_celsius(farenheit_temp):
    celsius = (farenheit_temp - 32)*5/9
    return celsius
````

First we used to `def` keyword to create the function definition. We labled the function `farenheit_to_celsius` to indicate that we are converting from farenheit to celsuis. The parameters of a function are then enclosed with parenthesis. The only parameter we are passing is a single value, the `farenheit_temp`. We convert the temperature to celsius and return the variable using the `return` keyword. 

````python
farenheit_to_celsius(100)
````

Let's convert the first five temperatures to celsius and display them side by side using the tools we have learned!

````python
for farenheit_temp in np_ohare_temperature[:5,1]:
    print(farenheit_temp,farenheit_to_celsius(farenheit_temp))
````
